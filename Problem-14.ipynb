Problem Statement - 14 | Longest Common Prefix

Write a function to find the longest common prefix string amongst an array of strings.
If there is no common prefix, return an empty string ""

**Approach** :- 


1) Take the first string as a reference.
2) Compare its characters with the characters at the same position in the other strings.
3) If a mismatch occurs or a string ends, return the prefix found so far.
4) If all characters match, return the entire first string as the prefix.


```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        pref = strs[0]
        pref_len = len(pref)

        for s in strs[1:]:
            while pref != s[0:pref_len]:
                pref_len -= 1
                if pref_len == 0:
                    return ""
                
                pref = pref[0:pref_len]
        
        return pref
```
