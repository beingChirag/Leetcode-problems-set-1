Problem Statement - 112 | Path Sum 

Given the root of a binary tree and an integer targetSum, return true if the tree has a root-to-leaf path such that the sum of all node values along the path equals targetSum.
A leaf node is a node with no children.

**Approach :-**

Use recursion to check all root-to-leaf paths.

1) Subtract the current node value from targetSum.

2) If we reach a leaf node and the remaining sum equals the node value, return True.

3) Recursively check the left and right subtrees.

```python
class Solution {
    public boolean hasPathSum(TreeNode root, int targetSum) {
        if (root == null) return false;

        if (root.left == null && root.right == null) {
            return targetSum - root.val == 0;
        }

        targetSum -= root.val;

        return hasPathSum(root.left, targetSum) || hasPathSum(root.right, targetSum);        
    }
}
```
