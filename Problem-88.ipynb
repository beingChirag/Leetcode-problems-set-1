Problem Statement - 88 | Merge Sorted Array

You are given two integer arrays nums1 and nums2 sorted in non-decreasing order, and two integers m and n.
Merge nums2 into nums1 as one sorted array.
The final sorted array should be stored inside nums1.
nums1 has a size of m + n, where:

1) The first m elements are valid numbers.
2) The last n elements are 0 and should be ignored.

**Approach :-**

1) Start merging from the end of both arrays.
2) Compare elements from nums1 and nums2.
3) Place the larger element at the end of nums1 and move the pointer backward.

```python
class Solution:
    def merge(self, nums1, m, nums2, n):
        i = m - 1
        j = n - 1
        k = m + n - 1

        while j >= 0:
            if i >= 0 and nums1[i] > nums2[j]:
                nums1[k] = nums1[i]
                i -= 1
            else:
                nums1[k] = nums2[j]
                j -= 1
            k -= 1
```
