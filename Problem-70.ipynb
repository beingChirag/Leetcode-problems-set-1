Problem Statement - 70 | Climbing Stairs 

You are climbing a staircase. It takes n steps to reach the top.
Each time you can climb 1 step or 2 steps.
Return the number of distinct ways to reach the top.

**Approach :-**

1) The number of ways to reach step n equals the sum of ways to reach n-1 and n-2.

2) This follows the Fibonacci pattern.

```python
class Solution {
    public int climbStairs(int n) {
        if (n <= 3) return n;

        int prev1 = 3;
        int prev2 = 2;
        int cur = 0;

        for (int i = 3; i < n; i++) {
            cur = prev1 + prev2;
            prev2 = prev1;
            prev1 = cur;
        }

        return cur;        
    }
}
```
