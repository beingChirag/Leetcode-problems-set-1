Problem Statement - 1 | TWO SUMS 


Given an array of integers numbers and an integer target, return the indices of the two numbers such that they add up to target.


Conditions :-

1) Each input has exactly one solution.

2) You cannot use the same element twice.

3) Return the answer in any order.


Approach (BRUTE FORCE) :-


1) Use two loops to check every pair of numbers.

2) For each pair, check if their sum equals the target.

3) If yes, return their indices.


```python
class Solution{
    public int[] twoSum(int[] nums, int target) {
        for(int i = 0; i < nums.length; i++) {
            for(int j = i + 1; j < nums.length; j++) {
                if(nums[i] + nums[j] == target) {
                    return new int[]{i, j};
                }
            }
        }
        return new int[]{};
    }
}
```
