Problem Statement - 26 | Remove duplicated from Sorted Array 

**Given an integer array numbers sorted in non-decreasing order, remove the duplicates in-place so that each unique element appears only once.
The relative order of the elements should remain the same.
Return the number of unique elements k.**

After removing duplicates:

1) The first k elements of numbers should contain the unique elements in sorted order.
2) The remaining elements beyond index k - 1 can be ignored.

**Approach : Use the two-pointer technique**

1) Traverse the array and keep track of the last unique element.
2) When a new unique number is found, place it next in the array.

```python
class Solution {
    public int removeDuplicates(int[] nums) {
        if (nums.length == 0) return 0;

        int i = 1;

        for (int j = 1; j < nums.length; j++) {
            if (nums[j] != nums[i - 1]) {
                nums[i] = nums[j];
                i++;
            }
        }

        return i;        
    }
}
```
