Problem Statement - 9 | Palindrome Number

Given an integer x, return true if x is a palindrome, and false otherwise.


We reverse the given number and compare it with the original number.

Steps:

1) Store the original number.

2) Reverse the digits of the number.

3) Compare the reversed number with the original number.

4) If both are equal → it is a palindrome.

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        s = str(x)
        return s == s[::-1]
```

Another Way :-

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        original = x
        reverse = 0
        
        if x < 0:
            return False
        
        while x > 0:
            digit = x % 10
            reverse = reverse * 10 + digit
            x = x // 10
        
        return original == reverse
```
