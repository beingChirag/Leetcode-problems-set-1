Problem Statement - 35 | Search Insert Position

Given a sorted array of distinct integers numbers and a target value, return the index if the target is found.
If the target is not found, return the index where it would be inserted in order.

**Approach :-**

1) Use Binary Search to find the target.
2) If the target is not found, return the position where it would be inserted.

```python
class Solution {
    public int searchInsert(int[] nums, int target) {
        int start = 0;
        int end = nums.length-1;

        while (start <= end) {
            int mid = start + (end-start)/2;
            if (nums[mid] == target) return mid;
            else if (nums[mid] > target) end = mid-1;
            else start = mid+1;
        }

        return start;
    }
}
```
