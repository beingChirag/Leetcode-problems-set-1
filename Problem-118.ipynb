Problem Statement - 118 | Pascal's Triangle 1

Given an integer numRows, return the first numRows of Pascal's Triangle.

**In Pascal's Triangle:**

1) Each number is the sum of the two numbers directly above it.

2) The first and last elements of every row are always 1.

**Approach :-**

Build the triangle row by row.

1) Start with the first row [1].

2) Each row begins and ends with 1.

3) For the middle elements, add the two numbers from the previous row.

```python
class Solution {
public:
    vector<vector<int>> generate(int numRows) {
        vector<vector<int>> result;
        for (int i = 0; i < numRows; i++) {
            vector<int> row(i + 1, 1);
            for (int j = 1; j < i; j++) {
                row[j] = result[i - 1][j - 1] + result[i - 1][j];
            }
            result.push_back(row);
        }
        return result;
    }
};
```
