Problem Statement - 67 | Add Binary 

Given two binary strings a and b, return their sum as a binary string.

**Approach :-**

1) Traverse both strings from the end, add corresponding digits with a carry, and build the result.
2) If a carry remains at the end, add it to the result.

```python
class Solution:
    def addBinary(self, a, b) -> str:
        x, y = int(a, 2), int(b, 2)
        while y:
            x, y = x ^ y, (x & y) << 1
        return bin(x)[2:]
```
