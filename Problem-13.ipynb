Problem Statement - 13 | Roman to Integer

Roman numerals are represented by seven different symbols:

| Symbol | Value |
|-------|-------|
| I | 1 |
| V | 5 |
| X | 10 |
| L | 50 |
| C | 100 |
| D | 500 |
| M | 1000 |

Roman numerals are usually written from largest to smallest from left to right.

Approach

1) Create a dictionary for Roman numeral values.

2) Traverse the string from left to right.

3) If the current value is less than the next value, subtract it.

4) Otherwise add it to the result.

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        res = 0
        roman = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }

        for a, b in zip(s, s[1:]):
            if roman[a] < roman[b]:
                res -= roman[a]
            else:
                res += roman[a]

        return res + roman[s[-1]] 
```
