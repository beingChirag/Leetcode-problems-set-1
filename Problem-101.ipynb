Problem Statement - 101 | Symmetric Tree

Given the root of a binary tree, check whether it is symmetric around its center.
A tree is symmetric if the left subtree is a mirror reflection of the right subtree.

**Approach :-**

Use recursion to compare the left and right subtrees.

1) If both nodes are None, return True.

2) If one node is None, return False.

3) If node values are different, return False.

**Recursively check:**

1) left child of left subtree with right child of right subtree

2) right child of left subtree with left child of right subtree.

3) If all mirror comparisons match, the tree is symmetric.

```python
class Solution {
public:
    bool isSymmetric(TreeNode* root) {
        return isMirror(root->left, root->right);
    }

private:
    bool isMirror(TreeNode* n1, TreeNode* n2) {
        if (n1 == nullptr && n2 == nullptr) {
            return true;
        }
        
        if (n1 == nullptr || n2 == nullptr) {
            return false;
        }
        
        return n1->val == n2->val && isMirror(n1->left, n2->right) && isMirror(n1->right, n2->left);
    }
};
```
