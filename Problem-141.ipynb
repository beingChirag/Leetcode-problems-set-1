Problem Statement - 141 | Linked List Cycle

Given head, the head of a linked list, determine if the list has a cycle.
A cycle occurs when a node’s next pointer points to a previous node, creating a loop in the list.

Return true if there is a cycle, otherwise return false.

**Approach :-**

1) Use two pointers (slow and fast).

2) slow moves one step, fast moves two steps.

3) If they meet, a cycle exists.

```python
public class Solution {
    public boolean hasCycle(ListNode head) {
        ListNode fast = head;
        ListNode slow = head;

        while (fast != null && fast.next != null) {
            fast = fast.next.next;
            slow = slow.next;

            if (fast == slow) {
                return true;
            }
        }

        return false;        
    }
}
```
