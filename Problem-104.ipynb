Problem Statement - 104 | Maximum Depth of Binary Tree

Given the root of a binary tree, return its maximum depth.
The maximum depth is the number of nodes along the longest path from the root node to the farthest leaf node.

**Approach :-**

Use recursion to calculate the depth of the tree.

1) If the node is None, return 0.

2) Recursively find the depth of the left subtree and right subtree.

3) The maximum depth is 1 + max(left_depth, right_depth).

```python
class Solution {
    public int maxDepth(TreeNode root) {
        if (root == null) {
            return 0;
        }
        return 1 + Math.max(maxDepth(root.left), maxDepth(root.right));        
    }
}
```
