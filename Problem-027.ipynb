Problem Statement - 27 | Remove Element

Given an integer array numbers and an integer val, remove all occurrences of val in-place.Return the number of elements k that are not equal to val.
The first k elements of numbers should contain the elements not equal to val. The remaining elements can be ignored.

**Approach :-**

Traverse array and copy every element not equal to val to the front of the array using a pointer.

```python
class Solution:
    def removeElement(self, nums, val):
        k = 0

        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1

        return k
```
