Problem Statement - 21 | Merge two sorted lists

Merge the two lists into one sorted linked list. The new list should be made by splicing together the nodes of the two lists.
Return the head of the merged linked list.

**Approach :-**


1) Compare values of list1 and list2.
2) Attach the smaller node to the merged list.
3) Repeat until one list becomes empty.
4) Attach the remaining nodes of the other list.

```python
class Solution {
    public ListNode mergeTwoLists(ListNode list1, ListNode list2) {

        if(list1!=null && list2!=null){
        if(list1.val<list2.val){
            list1.next=mergeTwoLists(list1.next,list2);
            return list1;
            }
            else{
                list2.next=mergeTwoLists(list1,list2.next);
                return list2;
        }
        }
        if(list1==null)
            return list2;
        return list1;
    }
}
```
