Problem Statement - 58 | Length of Last Word

Given a string s consisting of words and spaces, return the length of the last word in the string.
A word is defined as a substring consisting of non-space characters only.

**Approach :-**

Remove trailing spaces and count the characters of the last word until a space is encountered.

```python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        s=s.split()
        return len(s[-1])
```
