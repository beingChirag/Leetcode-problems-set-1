Problem Statement - 100 | Same Tree

Given the roots of two binary trees p and q, write a function to check whether they are the same tree.

Two binary trees are considered the same if:

1) They are structurally identical
2) The nodes have the same values

**Approach :-**

We use recursion to compare both trees.

1) If both nodes are None, return True.

2) If one node is None and the other is not, return False.

3) If the node values are different, return False.

Recursively compare:

1) left subtree of both trees

2) right subtree of both trees.

3) If all checks pass, the trees are the same.


```python
class Solution {
public:
    bool isSameTree(TreeNode* p, TreeNode* q) {
        if (!p && !q) {
            return true;
        }
        
        if (p && q && p->val == q->val) {
            return isSameTree(p->left, q->left) && isSameTree(p->right, q->right);
        }
        
        return false;        
    }
};
```
