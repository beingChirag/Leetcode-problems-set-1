Problem Statement - 119 | Pascal's Triangle 2

Given an integer rowIndex, return the rowIndex-th row (0-indexed) of Pascal's Triangle.

**In Pascal's Triangle:**

1) Each number is the sum of the two numbers directly above it.

2) The first and last elements of each row are always 1.

**Approach :-**

Build the row iteratively.

1) Start with the first row [1].

2) For each new row, update values from right to left.

3) Each element becomes the sum of itself and the previous element.

```python
class Solution(object):
    def getRow(self, rowIndex):
        row = [1]

        for _ in range(rowIndex):
            row = [left + right for left, right in zip([0]+row, row+[0])]
            
        return row    
```
