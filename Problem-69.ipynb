Problem Statement - 69 | Sqrt(x)

Given a non-negative integer x, return the square root of x rounded down to the nearest integer.
The returned integer should be non-negative.

**Approach :-**

1) Use Binary Search on the range 0 to x.

2) Find the largest number mid such that mid × mid ≤ x.

3) Return that value as the square root.

```python
class Solution(object):
    def mySqrt(self, x):
        if x < 2:
            return x
        
        i = 2
        while i * i <= x:
            i += 1
        
        return i - 1
        
```
